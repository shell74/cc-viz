```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pwd = os.getcwd()
```

```python
spotify = pd.read_csv(pwd + "/spotify_songs.csv")
```

```python
list(spotify.columns)
```

```python
spotify.describe()
```

```python
spotify.drop(['id', 'uri', 'track_href', 'analysis_url', 'song_name', 'Unnamed: 0', 'title', 'time_signature', 'type'], axis = 1, inplace = True)
```

```python
spotify.set_index('genre', inplace = True)
```

```python
plt.hist(spotify["danceability"], bins=30, color = 'blue')
```

```python
plt.hist(spotify["energy"], bins=20, color = 'blue')
```

```python
plt.scatter(spotify["danceability"],spotify["energy"], color = "red")
plt.title("Songs Energy vs Danceability")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.show()
```

```python
plt.figure(figsize = (13, 8))
sns.heatmap(spotify.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

```python
#seeing that there is a stronger correlation between energy of a song and loudness, 
# I want to investigate
```

```python
plt.figure(figsize = (13, 8)) #adjusts height and length of
plt.xticks(rotation=25) #adjusts the text in x to be rotated for visibility
sns.barplot(data=spotify, x=spotify.index, y='energy')
plt.show()
```

```python
plt.figure(figsize = (13, 8)) #adjusts height and length of
plt.xticks(rotation=25) #adjusts the text in x to be rotated for visibility
sns.barplot(data=spotify, x=spotify.index, y='loudness')
plt.show()
```

```python
# since loudness is recorded in negatives, the closer values to zero are considered less loud
# this shows that the relationship found in the matrix is the less loud the music, the higher the 
#energy for the song
```

```python
#lets take a look at a scatter plot comparing the two variables
fig, ax =plt.subplots(figsize=(15,7))
plt.scatter(spotify["loudness"],spotify["energy"], color = "purple")
plt.title("Songs Energy vs Loudness")
plt.xlabel("Loudness")
plt.ylabel("Energy")
plt.show()
```
